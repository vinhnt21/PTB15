- Yêu cầu người dùng nhập vào 1 số nguyên
- In ra : "Đây là số chẵn:"
    - Nếu n là số chẵn thì chỗ đằng sau là True
    - Nếu n không phải số chẵn thì chỗ đằng sau là False

- Ví dụ: người dùng nhập số 100:
    - In ra: "Đây là số chẵn: True"
- Ví dụ: người dùng nhập số 101:
    - In ra: "Đây là số chẵn: False"

```python
n = int(input('Nhập vào 1 số nguyên: '))
isEven = n % 2 == 0
print("Đây là số chẵn: ", isEven)
```
